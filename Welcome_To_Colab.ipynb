# Project 1 hotel **management**

```
chargesheet ={"room": "price",
              "AC room": 2000,
              "non AC room": 1500,
              "sweats": 15000,
              "dormetry": 1000,}
print("welcome to the Yashi's Resort")
print(chargesheet)
grand_total = 0
room_1 = input("Enter the required room you want to book = ")
if room_1 in chargesheet:
  grand_total += chargesheet[room_1]
  print(f"Your room {room_1} has been booked")
else:
  print(f"ordered room {room_1} is not available yet!")
another_booking = input("do you want to add another item? (yes/no) ")
if another_booking == "yes":
  room_2 = input("Enter the required room you want to book = ")
  if room_2 in chargesheet:
    grand_total += chargesheet[room_2]
    print(f"Your room {room_2} has been booked")
  else:
    print(f"ordered room {room_2} is not available")
print(f"the total amount to pay is {grand_total}")
```
