```
reduce(lambda x,y : x+y,[])
```

```
reduce(lambda x,y: x+y,[1])
```

## polymorphism#

```
class UPI:
  def pay(self):
    print("Payment through UPI")
class Card:
  def pay(self):
    print("Payment through Card")
for payment in [UPI(), Card()]:
  payment.pay()
```
