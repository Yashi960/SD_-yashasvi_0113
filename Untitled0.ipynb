```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter (+, -, *, /): ")

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
else:
    print("Invalid operator")
```

```python
def test_fib(n):
  a,b = 0,1
  for i in range(n):
    yield a
    a,b = b, a+b
```

```python
for i in test_fib(6):
  print(i)
```
